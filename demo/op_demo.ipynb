Data Pipelining with Op
-----------------------
Our goal here is to clean up and transform some scraped LinkedIn profiles in JSON format, so that we end up with profile-documents that we can use to fit a topic model. I'm just going to load up a small example subset of the JSON. 

```python
#=====[ here's some of the raw json. ]=====

linkedin_dict = json.load(open('./example_linkedin.json'))
pprint(linkedin_dict[:1]) 
```

```python
df = pd.DataFrame.from_dict(linkedin_dict)
df.set_index('name')
```

```python
#=====[ looks like some of those columns we don't need, or won't provide relevant info to the topic model. ]=====

linkedin_spurious_columns = [
    'certifications',
    'connections',
    'image',
    'image_last_modified',
    'image_sha1_base64',
    'last_modified',
    'url',
    'url_alt',
    'volunteering',
    'organizations',
    'projects',
    'summary-education',
    'employer'
]

#=====[ the main use case for ops is in chaining them together to prototype pipelines. ]=====
#=====[ but you can also use them as quick abstractions for common transforms on data: ]=====

pandas_formatting.drop_columns(cols=linkedin_spurious_columns) (df)
```

```python
#=====[ some of these columns are annoyingly formatted as lists of dicts with keys we don't care about. ]=====
#=====[ we'll use an Op to extract the values into flat lists. ]=====

print '\n======[ education ]=====\n'
pprint ( list( df['education'][:1] ))

print '\n======[ groups ]=====\n'
pprint ( list( df['groups'][:1] ))

print '\n======[ experience ]=====\n'
pprint ( list( df['experience'][:1] ))
```

```python
#=====[ looks like "current" has some interesting stuff, but it's all singleton lists of strings. ]=====
#=====[ we'll operate on that column and extract the singleton element. ]=====


print '\n======[ current ]=====\n'
pprint( list(df['current'][:3]))
```

```python
#=====[ a bunch of columns are now going to be lists of strings. we'll just join them into one long string. ]=====

df[['skills', 'past']]
```

```python
#=====[ sometimes there's a series of Ops that we find ourselves using together often. ]=====
#=====[ chaining them together is just "+" ]=====

preprocess_raw_text = \
    text_preprocessing.str_fillna +\
    text_preprocessing.str_remove_nonalpha +\
    text_preprocessing.str_lower +\
    text_preprocessing.str_split
```

```python
#=====[ let's define an op to get the useful text out of the dict. ]=====
#=====[ we can just use the @op decorator.                         ]=====


@op(num_out=1)
def extract_list_dict_values(df, cols=None, key=None):
    """
        usage: 
        ------
            extract_list_dict_values(cols=['a'], key="'b'") (df)
        params:
        -------
            - col: column containing a list of dicts for each entry
            - key: key of dict value to extract
        description:
        ------------
            outputs into column named {incol}_{key} the result of extracting the values 
            described by 'key' in each element of the corresponding entry in 'incol'
    """
    for col in cols:
        df["{incol}_{key}".format(incol=col, key=key)] = df[col].apply(lambda l: [d[key] for d in l] if type(l) == list and type(l[0]) != str else np.nan)
    return df

df = extract_list_dict_values (cols=['education', 'groups'], key="'title'") (df)
df = extract_list_dict_values (cols=['education'], key="'details-education'") (df)
df = extract_list_dict_values (cols=['experience'], key="'summary'") (df)
```

```python
#=====[ here's an example of how we can string Ops together into a complete preprocessing chain. ]======

linkedin_preprocess_chain = \
    pandas_formatting.drop_columns (cols=linkedin_spurious_columns) +\
    text_preprocessing.str_join (cols=['current', 'skills', 'past', 'experience', 'groups', 'education']) +\
    preprocess_raw_text( cols=[ #=====[ Textual Columns ]=====
        'name',
        'industry', 
        'title', 
        'education',
        'experience',
        'groups', 
        'current', 
        'past', 
        'skills',
        'location',
    ]) +\
    text_preprocessing.join_text_columns (incols=['current', 'past', 'industry', 'title', 'groups', 'skills'], outcol=['user_doc']) +\
    text_preprocessing.remove_stopwords (cols=['user_doc'], langs=['english','german','norwegian','swedish','danish']) +\
    text_preprocessing.dedupe_list (cols=['user_doc'])
```

```python
df = linkedin_preprocess_chain (df)
```

```python
#======[ now we've got a "user_doc" column that has the useful tokenized list of words for each profile.     ]======
#======[ time to train a model! i'll throw it into an LDA model with gensim, but there are too few documents ]======
#======[ to get topics that are actually good. it's just an illustration of the concept though.              ]======

df['user_doc'][2]
```

```python
gensim_dict = topic_modelling.dictionary_fit(col="'user_doc'")(df)
```

```python
pprint(gensim_dict.token2id)
```

```python
df = topic_modelling.dictionary_transform(col='"user_doc"') (df, gensim_dict)
```

```python
#=====[ the bag of words sparse representation is now in the user_doc_bow column ]=====

df.user_doc_bow
```

```python
model = topic_modelling.lda_fit(col='"user_doc"', num_topics=20) (df, gensim_dict)
```

```python
pprint(model.show_topics())
```
